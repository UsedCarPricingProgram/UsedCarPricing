```python
#导入相关包
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
%matplotlib inline
```

```python
df = pd.read_excel('/Users/heng.w/Desktop/cleaned_car_all.xlsx')

df.head(10)
```

```python
df.dtypes
```

```python
df.describe()
```

```python
#将新车价格对数化
df['price_log'] = df['price_new'].apply(np.log1p)

#增加「rate_residual」
df['rate_residual'] = df['price']/df['price_new']

#计算「使用天数」
now=datetime(2022,11,8)
df['date_regi'] = pd.to_datetime(df['date_regi'], format='%Y-%m')
df['num_day']=df['date_regi'].apply(lambda x: now-x)
df['num_day']=df['num_day'].apply(lambda x: x.days)
df.head(10)
```

```python

sns.set_style('ticks')
fig, ax = plt.subplots(1,3,constrained_layout=True, figsize=(16, 4), dpi=180)
#不同gearbox的保值率分布
axesSub = sns.violinplot(data=df,y='rate_residual',x='gearbox',split=True,ax=ax[0])
axesSub.set_title('rate_residual distribution of gearbox')
#不同num_trans的保值率分布
sns.violinplot(data=df,y='rate_residual',x='num_trans',split=True,ax=ax[1])
axesSub.set_title('rate_residual distribution of num_trans') 
#不同energy_type的保值率分布
sns.violinplot(data=df,y='rate_residual',x='energy_type',split=True,ax=ax[2])
axesSub.set_title('rate_residual distribution of energy_type')
```

```python
#不同license_location的保值率分布
sns.set_style('ticks')
fig, ax = plt.subplots(1,3,constrained_layout=True, figsize=(16, 4), dpi=180)
axesSub = sns.violinplot(data=df,y='rate_residual',x='license_location',split=True,ax=ax[0])
axesSub.set_title=('rate_residual distribution of license_location')
#不同color的保值率分布
axesSub = sns.violinplot(data=df,y='rate_residual',x='color',split=True,ax=ax[1])
axesSub.set_title=('rate_residual distribution of color')
axesSub = sns.violinplot(data=df,y='rate_residual',x='standard',split=True,ax=ax[2])
axesSub.set_title=('standard distribution of color')
```
