```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import os
import re
```

```python
#读取清洗后的数据
clean_data_petrol = pd.read_excel(os.path.join(os.getcwd(),'../data/cleaneddata/cleaned_car_all.xlsx'),sheet_name='petrol')
clean_data_electric = pd.read_excel(os.path.join(os.getcwd(),'../data/cleaneddata/cleaned_car_all.xlsx'),sheet_name='electric')
clean_data_mixed = pd.read_excel(os.path.join(os.getcwd(),'../data/cleaneddata/cleaned_car_all.xlsx'),sheet_name='mixed')
```

```python
def show_corr(clean_data_set):
    corr_data_matrix = clean_data_set.corr()
    corr_data_matrix['price'].sort_values(ascending=False)
    #查看哪些字段与汽油车价格最相关，我们先计算相关性矩阵
    df_tep1=clean_data_set.drop(columns='energy_type')
    sns.set(rc={"figure.figsize":(20, 20)})
    sns.heatmap(df_tep1.corr(), annot = True)

show_corr(clean_data_petrol)
```

通过上述的热力图可以发现与价格相关系数大的有price_new,official_endurance,gearbox,battery_capacity,displacement....

```python
show_corr(clean_data_electric)
```

```python
show_corr(clean_data_mixed)
```
